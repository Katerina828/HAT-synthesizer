```python
def choose_disjoint(fake_sample,training,soft_list,tanh_list,device):
    temp = fake_sample[:,tanh_list[0]:(tanh_list[-1]+1)]
    for i in range(len(soft_list)-1):
        tem_data = fake_sample[:,soft_list[i]:soft_list[i+1]]
        x = torch.zeros_like(tem_data,device = device)
        index = torch.argmax(tem_data,dim=1)
        x[np.arange(0,fake_sample.shape[0],1),index]=1
        temp = torch.cat([temp,x],1)
    tem_data = fake_sample[:,soft_list[-1]:]
    x = torch.zeros_like(tem_data,device = device)
    tem_soft = torch.argmax(tem_data,dim=1)
    x[np.arange(0,fake_sample.shape[0],1),tem_soft]=1
    temp = torch.cat([temp,x],1)
    
    #transform age
    a = torch.tensor([73.])
    a = a.repeat(fake_sample.shape[0],1)
    b= torch.ones([fake_sample.shape[0],fake_sample.shape[1]-1])
    c = torch.cat([a,b],1).to(device)
    
    d = torch.tensor([19.])
    d = d.repeat(fake_sample.shape[0],1)
    e= torch.zeros_like(b)
    f = torch.cat([d,e],1).to(device)
    
    temp = torch.round(temp* c + f)
    
    a = torch.tensor([73.])
    a = a.repeat(training.shape[0],1)
    b= torch.ones([training.shape[0],training.shape[1]-1])
    c = torch.cat([a,b],1).to(device)
    
    d = torch.tensor([19.])
    d = d.repeat(training.shape[0],1)
    e= torch.zeros_like(b)
    f = torch.cat([d,e],1).to(device)
    
    
    training = torch.round(training*c +f)
    
    concat = torch.cat((temp,training), axis=0)
    unique,inverse_index,count = torch.unique(concat,dim=0, return_inverse=True,return_counts=True)
    x = ~(count-1)[inverse_index][:temp.shape[0]].bool()
    choose_index = torch.nonzero(x).squeeze()
    
    return choose_index
```

```python
def MI_attack(Discriminator,D_in,D_out):
    D_in = torch.from_numpy(D_in[0:1000].astype('float32')).to(gan.device)
    D_out = torch.from_numpy(D_out[0:1000].astype('float32')).to(gan.device)

    D_in_pred = Discriminator(D_in)
    D_out_pred = Discriminator(D_out)

    D1 = torch.cat([D_in,D_in_pred],dim =1)
    D2 = torch.cat([D_out,D_out_pred],dim =1)

    label_in = torch.ones(1000,1).to(gan.device)
    label_out = torch.zeros(1000,1).to(gan.device)

    D1 = torch.cat([D1,label_in],dim =1).detach().cpu().numpy()
    D2 = torch.cat([D2,label_out],dim =1).detach().cpu().numpy()
    dataset = np.concatenate([D1,D2],axis=0)
    return dataset
dataset = MI_attack(myD,gan.ganinput,gan.testinput)
```
